```python
def hasvisitedallinlist(df, list):
    for i in range(len(list)):
        df_cat = df[df['cat'] == list[i]]
        if i == 0:
            uids = df_cat['hash_uid'].value_counts().index.tolist()
        else:
            ul = len(uids)
            for j in range(len(df_cat['hash_uid'].value_counts().index.tolist())):
                if df_cat['hash_uid'].value_counts().index.tolist()[j] in uids:
                    uids.append( df_cat['hash_uid'].value_counts().index.tolist()[j])
            uids = uids[ul:]
            # uidsset = set(newuids)
            # uids = sorted(uidsset)

    return uids


                    
```

```python
def totextvisited(dataframe, parentheses):
    
    df = dataframe.copy()
    df.insert(0, 'date', df['local_datehour'].str.split(" ", expand = True)[0])
    df.insert(1, 'hour', df['local_datehour'].str.split(" ", expand = True)[1])
    df = df.drop(['local_datehour'], axis = 1)
    df['tag'] = df['tag'].replace(np.nan, "")
    df['placename'] = df['placename'] + " " + df['tag'] 
    df = df.drop(['tag'], axis = 1)
    df = df.sort_values(by = ['date', 'hour'])
    
    print("A person's visit history from " + df.iloc[0, 0][5:10] + df.iloc[0, 0][4] + df.iloc[0, 0][0:4] 
          + " to " + df.iloc[len(df) - 1, 0][5:10] + df.iloc[len(df) - 1, 0][4] + df.iloc[len(df) - 1, 0][0:4] + " is as follows:\n")
    
    visited = {}
    visitedcataddr = {}
    
    for i in range(len(df)):
        irow = df.iloc[i, :]
        
        if irow['placename'][-1] == " ":
        
            pn = str(irow['placename'][:-1])
            
        else:
            
            pn = str(irow['placename'])
            
        if pn in visited.keys(): 
            
            visited[pn] += ", at " + irow['hour'] + " on " + irow['date'][5:10] + irow['date'][4] + irow['date'][0:4]
            
        elif pn not in visited.keys():
            
            visited.update({pn : " at " + irow['hour'] + " on " + irow['date'][5:10] + irow['date'][4] + irow['date'][0:4]})
            visitedcataddr.update({pn : [irow['cat'].lower(), str(irow['pid']), irow['addr'], 
                                        irow['lat_lng'].split(",")[0], irow['lat_lng'].split(",")[1]]})
            
    if parentheses == False:
        
        print("The person visited", end = "")
        
        for key, value in visited.items():
            if key == list(visited.keys())[-1]:
                print(" a " + visitedcataddr[key][0] + " named " + key 
                        + " located at " + visitedcataddr[key][2] + value + ".")
            else:
                print(" a " + visitedcataddr[key][0] + " named " + key 
                        + " located at " + str(visitedcataddr[key][2]) + value + ",", end = "")
                
    else:
        
        print("The person visited", end = "")
        
        for key, value in visited.items():
            if key == list(visited.keys())[-1]:
                print(" a " + visitedcataddr[key][0] + " named " + key + " (pid: " + visitedcataddr[key][1]
                        + ") located at " + visitedcataddr[key][2] 
                        + " (latitude: " + visitedcataddr[key][3] + ", longitude: " + visitedcataddr[key][4] + ")"
                        + value + ".")
            else:
                print(" a " + visitedcataddr[key][0] + " named " + key + " (pid: " + visitedcataddr[key][1]
                        + ") located at " + visitedcataddr[key][2] 
                        + " (latitude: " + visitedcataddr[key][3] + ", longitude: " + visitedcataddr[key][4] + ")"
                        + value + ",", end = "")
            
```
